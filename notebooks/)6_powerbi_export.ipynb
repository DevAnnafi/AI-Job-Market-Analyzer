```python
import pandas as pd
import os

df = pd.read_csv("../notebooks/ds_salaries_features_v1.csv")
print(df.columns.tolist())

os.makedirs("../powerbi_data", exist_ok=True)
```

```python
salary_trends = (
    df.groupby(["work_year", "role_category"])["salary_in_usd"]
      .mean()
      .reset_index()
)
salary_trends.columns = ["work_year", "role", "avg_salary_usd"]
salary_trends.to_csv("../powerbi_data/salary_trends.csv", index=False)
```

```python
remote_trends = (
    df.groupby(["work_year", "role_category"])["remote_ratio"]
      .mean()
      .reset_index()
)
remote_trends.columns = ["work_year", "role", "remote_ratio"]
remote_trends.to_csv("../powerbi_data/remote_trends.csv", index=False)
```

```python
country_salary = (
    df.groupby("company_location")["salary_in_usd"]
      .mean()
      .reset_index()
)
country_salary.columns = ["company_location", "avg_salary_usd"]
country_salary.to_csv("../powerbi_data/country_salary.csv", index=False)
```

```python
experience_salary = (
    df.groupby("experience_level")["salary_in_usd"]
      .mean()
      .reset_index()
)
experience_salary.columns = ["experience_level", "avg_salary_usd"]
experience_salary.to_csv("../powerbi_data/experience_salary.csv", index=False)
```
